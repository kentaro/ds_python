その前に、「NaNって何だ？」について真面目にみていこう。NaNってのは、Not a Numberの略ってのは第9回でやりましたね。

これからいろんなデータを扱っていく上で、NaNをどのように処理するかってのは適切な分析をする上で重要なことなのです！実際に扱うデータは、必ずしも全部のレコードに値がぎっしりつまっているとは限りません。実際、これまで使ってきたタイタニックのデータにもNaNがいっぱいあるんですね〜。

ではいつものやつ〜〜〜。

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/titanic/train.csv')
df
```

`Cabin`カラムにたくさんNaNがありますね〜。なんでなんだろ？客室がなくてそのへんに雑魚寝でもしてたのかな……。

というわけで、このNaNに対してどう扱っていくかってのを見ていきましょう！

まずは、NaNのあるレコードをdropしちゃう、`df.dropna()`！

```python
df.dropna()
```

もともと891行あったデータが183行にまで減っちゃいましたね！ちなみに、第13回でやったフィルタリングのときと同じように、dropしたあともindexは維持されてますね。

データがたくさんある場合は`df.dropna()`でバーンと落としちゃってもいいかもですが、183行しか残らないようだと不安ですね……。

さて、`df.dropna()`に引数`axis=1`をわたすとNaNを含むカラムをdropできます！（他のメソッドと同じく、`axis=1`とするとカラムを表すんですね）。

```python
df.dropna(axis = 1)
```

でもこれだと一個でもNaNがあると説明変数を削っちゃうことになるので、慎重に考えないとならないみたいです！（理由は、かめさんがのちに機械学習入門で教えてくれるそうです。楽しみですね！）

また、指定したカラムでNaNがあるレコードだけを落とす方法もあります。

```python
df.dropna(subset=['Age'])
```

ちょっとわかりにくいですが、indexが888のレコードが、`Age`カラムの値がNaNだったのでdropされてます！

さて、これまではNaNなデータをdropしちゃうって方法を見てきましたが、なにか別の値を入れる方法もあります。

たとえば`df.fillna()`を使うと、一律に値をいれられます。

```python
df.fillna('THIS IS IT!!1')
```

でも、実際の現場ではNaNを一律に何かにするみたいなことはないそうですね……。その代わり、何か代表値を入れることが多いんですね〜〜…たとえば、`Age`なら平均年齢とか。

まず、あるカラムの平均値はこうやって取ります。

```python
df['Age'].mean()
```

ほんでもってこの平均値を`Age`カラムのNaNなとこに入れたいんだけど、どうするか？

まずはSeriesを作りましょう〜。

```python
df['Age'].fillna(df['Age'].mean())
```

888の行に、さっき計算した平均年齢が入ってますね。ここは、元々NaNだったところです。そして、このSeriesを元のDataFrameに突っ込んでやると〜〜〜？

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

ジャジャーン！元のDataFrameの`Age`カラムが上書きされて、NaNだった888のレコードに平均年齢が入力されました〜〜〜。パチパチパチ。

さて、DataFrameの中の値がNaNかどうかを調べるにはどうしたらよいでしょうか？

NumPyには`np.isnan()`ってのがあるんですが、Pandasだと`pd.isna()`を使います（最後の`n`がないのがポイント）。

```python
df = pd.read_csv('data/titanic/train.csv')
pd.isna(df)
```

NaN判定の結果を別カラムで持ちたいときはこう！

```python
df['Cabin_nam'] = pd.isna(df['Cabin'])
df
```

一番右に、`Cabin_nan`というカラムができて、`True`/`False`が入りましたね〜。このように、新たにカラムを作りたいときは

`df['新しいカラム名'] = Series`とすると、もしそのカラムがあれば代入するし、なければ新規に作られるんですね。便利！
