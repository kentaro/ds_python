今日もDataFrameに対するフィルタ操作をみていきましょう〜〜〜。

さて、いつものやつ〜〜〜。

```python
import pandas as pd
```

今回も、前回同様タイタニックのデータを使いま〜す。

```python
df = pd.read_csv('data/titanic/train.csv')
df
```

さて、この891レコードから、生存者のレコードのみを抽出してみましょう。

まずは、カラムが`Survived`なSeriesを取り出してみましょう。

```python
df['Survived']
```

死亡は`0`、生存は`1`で記録されているようですね。

今度は次のようにしてみます。

```python
df['Survived'] == 1
```

カラムの値が、`0`/`1`から`True`/`False`に変わりましたね。どういうことなんでしょうか？

```python
print(type(df['Survived']))
print(type(df['Survived'] == 1))
```

同じ型が返ってきますね。きっと、`==`演算子をオーバーロードして、レシーバーのSeriesの指定されたカラムを一個ずつ検証していって`True`/`False`のSeriesとして返すみたいな感じなんでしょうね。

さて、その`df['Survived'] == 1`を`df[]`に渡してみると……

```python
df[df['Survived'] == 1]
```

`Survived`が`1`のデータだけが返ってきた〜〜〜！不思議な挙動だなあ。きっと、`df['Survived'] == 1`には、`Survived`カラムをフィルタしてね、っていう情報も入ってるんでしょうね〜〜〜。

ここはやっぱり、生存者のみを抽出したこのDataFrameの統計量を見てみくなりますよね！そんなときは……

```python
df[df['Survived'] == 1].describe()
```

ジャジャーン！元のDataFrameと比較してみましょう。

```python
df.describe()
```

`Fare`（料金）の平均値（`mean`）を見ると、生存者のほうがちょっと高いですね……。これが意味するところは……と興味がつきないところですが、この差異って本当に差があるといえるの？ってことについては、かめさんが統計学入門講座で解き明かしてくれるそうです！楽しみですね！

さて、他にもいろんな条件でフィルタしてみますか〜〜〜。

```python
df[df['Age'] >= 60]
df[df['Pclass'] == 1]
df[df['Sex'] == 'female']
```

こんな感じでいろいろ抽出できちゃいます！

さて、こうなってくると今度は、抽出条件をあれこれ組み合わせたくなりますね！

Python（やSQLとか）だと`and`や`or`を使うところですが、DataFrameのフィルタ操作は`&`と`|`を使うんですね〜。注意です！

60歳以上の女性の場合はこんな感じ。

```python
df[(df['Age'] >= 60) & (df['Sex'] == 'female')]
```

1stクラスの乗客、もしくは、10歳未満の場合！

```python
df[(df['Pclass'] == 1) | (df['Age'] < 10)]
```

いっぱいでてきたな……。

さて、いままでは二項演算子を見てきましたが、NOT演算を表す単項演算子もあるんですね〜。それが`~`！条件の真偽が逆転します。値がbooleanのカラムでフィルタするときによく使うんだそうですね！

たとえばこんなデータがあるとしましょう。

```python
data = [{'Name':'John', 'Survived':True},
        {'Name':'Emily', 'Survived':False},
        {'Name':'Ben', 'Survived':True}]
df = pd.DataFrame(data)
df
```

生存者だけでフィルタしてみます。

```python
df[df['Survived']]
```

さて、`Survived == False`のときはどうでしょう？そんな時に`~`を使うんですね〜。それっ。

```python
df[~df['Survived']]
```

ところでですね、勘のいい方はすでにお気づかもしれません。さっきのタイタニックのデータフィルタしたDataFrameをもう一度見てみましょうか。

```python
df = pd.read_csv('data/titanic/train.csv')
df[df['Survived'] == 1]
```

わかりましたか？わかりませんかね？実はですね、indexがとびとびになってるところがあるんですね〜。これは、元のDataFrameのindexが保存されてるわけですね。

これに対してindexを振り直したいというときは`df.reset_index()`を使いましょう。

```python
df.reset_index()
```

そうすると、元のindexが`index`カラムとして保存されつつ、新たにindexが振られるわけですね〜〜〜。

もひとつついでに、別のカラムをindexにしたい場合は`df.set_index()`を使います。`Name`カラムをindexにしてみましょう。

```python
df = pd.read_csv('data/titanic/train.csv')
df.set_index('Name')
```

indexを貼り直すってことはよくあるみたいです。でも、せっかく覚えたんですが、indexをセットすることはあんまりないみたいですね〜〜〜。
