この時、↑のようにindex情報が`Unnamed:0`として保存されるが、これはだいたい必要ないので、`index=False`を渡して消しちゃう。

```python
df = pd.read_csv('data/titanic/train.csv')
df['Adult'] = df['Age'].apply(lambda x: x > 20)
df.to_csv('data/titanic/train_w_adult.csv', index=False)
```

```python
df = pd.read_csv('data/titanic/train_w_adult.csv')
df.head(3)
```

## `.itterrows()`でDataFrameをイテレーション

```python
df = pd.read_csv('data/titanic/train.csv')
for idx, row in df.iterrows():
    if row['Age'] > 40 and row['Pclass'] == 3 and row['Sex'] == 'male' and row['Survived'] == 1:
        print('{} is very lucky guy...!'.format(row['Name']))
```

前回の`.apply()`は各レコードの処理をした結果を別のカラムに保存するときに使い、今回の`.iterows()`は値を返すのではなく処理だけをしたいときに使うことが多い。

## `.sort_values()`で特定のカラムでソート

たとえば、年齢が若い順にソートするみたいな場合。

```python
df.sort_values('Age')
```

デフォルトは昇順なので、皇潤の場合は`ascending=False`を指定する。

```python
df.sort_values('Age', ascending=False)
```
